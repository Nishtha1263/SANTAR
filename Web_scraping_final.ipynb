```python
workbook_name = f'stop2-fra.xlsx'    #Change 

with pd.ExcelWriter(workbook_name, engine='xlsxwriter') as writer:
        curr_url = f'https://www.worldaquatics.com/competitions/308/fina-airweave-swimming-world-cup-2015/results'    #Change
        driver.get(curr_url)

        buttons = driver.find_elements(By.CLASS_NAME, 'schedule__item')
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        titles = soup.find_all('li', class_='schedule__item')  

        # Iterate over each button element
        for i, button in enumerate(buttons):
            try:
                    button.click()
            except:
                     pass
            
            time.sleep(2)

            heat_buttons = driver.find_elements(By.CLASS_NAME, 'unit-selector__unit')
            wait = WebDriverWait(driver, 10)
            h_button = wait.until(EC.element_to_be_clickable((By.CLASS_NAME, 'unit-selector__unit')))

            # Iterate over first two heat buttons
            for h_button in heat_buttons[:2]:
                try:
                    h_button.click()
                except:
                     pass
                time.sleep(2)

                soup = BeautifulSoup(driver.page_source, 'html.parser')
                head = soup.find('button', class_='unit-selector__unit-select js-results-unit is-active')
                name_of_heat = head['data-unit-name']

                table_rows_even = soup.find_all('tr', class_='results-table__row js-results-table-row')
                table_rows_odd = soup.find_all('tr', class_='results-table__row results-table__row--odd js-results-table-row')
                table_rows = table_rows_even + table_rows_odd

                df = pd.DataFrame(columns=['Rank', 'Lane', 'Country', 'FN', 'LN', 'Country2', 'Age', 'RT', 'Time', 'Time Behind', 'Points'])

                for tr in table_rows:
                    # Split the text and store it in info_player
                    info_player = tr.get_text().split()

                    # Check if info_player has at least 11 elements before accessing them
                    if len(info_player) == 11:
                            df = df.append({'Rank': info_player[0], 
                                            'Lane': info_player[1], 
                                            'Country': info_player[2], 
                                            'FN': info_player[3], 
                                            'LN': info_player[4], 
                                            'Country2': info_player[5], 
                                            'Age': info_player[6],
                                            'RT': info_player[7], 
                                            'Time': info_player[8], 
                                            'Time Behind': info_player[9], 
                                            'Points': info_player[10]}, ignore_index=True)
                    elif len(info_player) == 12:
                            df = df.append({'Rank': info_player[0], 
                                    'Lane': info_player[1], 
                                    'Country': info_player[2], 
                                    'FN': info_player[3], 
                                    'LN': info_player[4], 
                                    'Country2': info_player[6], 
                                    'Age': info_player[7],
                                    'RT': info_player[8], 
                                    'Time': info_player[9], 
                                    'Time Behind': info_player[10], 
                                    'Points': info_player[11]}, ignore_index=True)
                    elif len(info_player) == 13:
                        df = df.append({'Rank': info_player[0], 
                                    'Lane': info_player[1], 
                                    'Country': info_player[2], 
                                    'FN': info_player[3], 
                                    'LN': info_player[4], 
                                    'Country2': info_player[7], 
                                    'Age': info_player[8],
                                    'RT': info_player[9], 
                                    'Time': info_player[10], 
                                    'Time Behind': info_player[11], 
                                    'Points': info_player[12]}, ignore_index=True)
                    elif len(info_player) < 11:
                        continue
                df.drop(columns=['Country2'], inplace=True)

                # Store name_of_heat in the worksheet name
                sheet_name = f"{titles[i].get_text().strip()}-{name_of_heat}"
                # Check if the length of the sheet name exceeds the maximum allowed length (31 characters)
                if len(sheet_name) > 31:
                    # Truncate the sheet name to fit within the maximum length
                    sheet_name = sheet_name[:31]


                # Write the dataframe to the Excel worksheet
                df.to_excel(writer, sheet_name=sheet_name, index=False)

                time.sleep(1)
```
